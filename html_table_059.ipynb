# Get HTML Table to DataFrame
Data in tabular formats, such as CSV files, is very easy to process because the data relationships are straightforward. HTML tables are another format that allows us to easily analyze data. Copying or downloading HTML tables can be more difficult than you might think, as they often come with CSS styles that make reading the data difficult. Fortunately, the pandas library provides tools that allow us to easily extract data from HTML tables and place them in a DataFrame.

The following code example reads all HTML tables exist on the webpage at the specified URL as a list of dataframe.

```python
import pandas as pd
import requests
from io import StringIO

url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population_(United_Nations)"

# Add headers to mimic a web browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Fetch the HTML content using requests
try:
    response = requests.get(url, headers=headers)
    response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
    html_content = response.text

    # Read all tables from the webpage content
    # Wrap html_content in StringIO to suppress the FutureWarning
    tables = pd.read_html(StringIO(html_content))

    # tables is a list of DataFrames
    print(f"Number of tables found: {len(tables)}")

    # Get the first table
    df = tables[0]

    print(df.head())
except requests.exceptions.RequestException as e:
    print(f"Error fetching URL: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

Note: The code above might not work because modern web pages do not allow such a script to run. Use your preferred AI assistant to correct the code.

## Data cleaning
The retrieved data usually needs to undergo a cleaning process. The data cleaning process often involves assigning appropriate header names. Long and difficult-to-spell header names can negatively impact coding efficiency.

### Exercise 2
Change column names to `Country`, `Jul2022`, `Jul2023`, `Change`, `Region`, `Subregion`.

```python
# Exercise 2
# This block should print the dataframe with new column names
new_column_names = {
    'Country or territory': 'Country',
    'Population (1 July 2022)': 'Jul2022',
    'Population (1 July 2023)': 'Jul2023',
    'Change (%)': 'Change',
    'UN continental region[1]': 'Region',
    'UN statistical subregion[1]': 'Subregion'
}
df = df.rename(columns=new_column_names)
display(df.head())
```

Often, tables contain unwanted characters such as references and footnotes. This is very common when scraping Wikipedia tables. You typically see things like:
- China[a]
- Tanzania[b]
- France[c]
We can apply `str.replace()` with `regex` to remove these unwanted characters.

### Exercise 3
Remove footnotes ([a], [b], [c], ...) and colonization ((United States), (France), (China), ...) from the 'Country' column. You may use AI assistant. After removing:
- *China[a]* should appear as *China*
- *Tanzania[b]* should appear as *Tanzania*
- ...
- *Hong Kong (China)* should appear as *Hong Kong*
- *Puerto Rico (United States)* should appear as *Puerto Rico*

```python
# Exercise 3
# This block should clean the 'Country' column by removing footnotes and parenthetical information

# Remove text within square brackets (footnotes)
df['Country'] = df['Country'].str.replace(r'\[.*?\]', '', regex=True)

# Remove text within parentheses and the parentheses themselves (e.g., '(China)')
# This regex specifically targets ' (CountryName)' patterns that typically follow the main country name
df['Country'] = df['Country'].str.replace(r'\s*\(.*?\)', '', regex=True)

# Trim any leading/trailing whitespace that might be left after removal
df['Country'] = df['Country'].str.strip()

display(df.head())
```

Important: The first important step before applying data cleaning is to observe the data and see if any cleaning can be applied. For example, using regex to remove all parentheses content may also remove other meaningful info.
